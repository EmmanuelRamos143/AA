# Investigacion Clustering Espectral

En años recientes, el análisis espectral de clusters ha sido una técnica popular de clustering ya que este se desempeña en algunos casos mejor que los algoritmos de clustering tradicional tales como k-means.

Los tres principales paso relacionados a spectral clustering son, construir un gráfico de similitud, proyectar los datos en espacios de dimensión menores, aplicar clustering a los datos.
Los puntos para elaborar el algoritmo serian:
1.	Formar una matriz de distancias
2.	Transformar la matriz de distancias en una matriz de Afinidad ‘A’.
3.	Calcular el grado de la matriz y la matriz Lapacian **L = D-A**
4.	Encontrar los eigenvalores y eigenvectores de L
5.	Con los valores dados de los eigenvectores de los K mayores eigenvalores del paso anterior
6.	Normalizar los vectores
7.	Clusterizar los puntos de datos en k-dimensiones espaciales

¿Pros y contras al usar Espectral Clustering?

Comparado con otros algoritmos es **computacionalmente más rápido que otros algoritmos** ya que distribuye el dataset en miles de puntos de datos.
Una de las contras en usar este algoritmo es que ***el calculo de las distancias podría ser costoso de computar ya que primero se tienen que calcular sus eigenvectores y eigenvalores***.

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
from numpy import random
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(heart_df)
```

```python
from sklearn.preprocessing import StandardScaler, normalize 
from sklearn.decomposition import PCA 
from sklearn.metrics import silhouette_score 
import pandas as pd 
import matplotlib.pyplot as plt 
# Normalizing the Data 
X_scaled = scaled
X_normalized = normalize(X_scaled) 
  
# Converting the numpy array into a pandas DataFrame 
X_normalized = pd.DataFrame(X_normalized) 
  
# Reducing the dimensions of the data 
pca = PCA(n_components = 2) 
X_principal = pca.fit_transform(X_normalized) 
X_principal = pd.DataFrame(X_principal) 
X_principal.columns = ['P1', 'P2'] 
  
X_principal.head(2)

##Use PCA first 
spectral_model_rbf = SpectralClustering(n_clusters = 2, affinity ='rbf') 
  
# Training the model and Storing the predicted cluster labels 
labels_rbf = spectral_model_rbf.fit_predict(X_principal)
```

```python
# Visualizing the clustering 
plt.figure(figsize=(10, 8), dpi=80)
plt.title("Spectral Clustering")
plt.scatter(X_principal['P1'], X_principal['P2'],  
           c = SpectralClustering(n_clusters = 2, affinity ='rbf') .fit_predict(X_principal), cmap =plt.cm.winter) 
plt.show() 
```

Cnclusiones el Clustering Espectral es rapido al ejecutarse en el dataset y se ve claramente una division de los datos de manera adecuada.
